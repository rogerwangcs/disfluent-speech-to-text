```python
import os
import librosa
import speech_recognition as sr
import IPython.display as ipd
import soundfile as sf

import warnings
warnings.filterwarnings('ignore')
```

## Make Predictions

```python
# audio_dir = './librispeech-big-lab'
# predictions_dir = './librispeech-big-lab-preds'
audio_dir = './augmented-data-big'
predictions_dir = './augmented-data-big-preds'
test_speakers = ['78', '302', '248']

# replace a filename string's extension with a new extension
def replace_extension(file, ext):
    return os.path.splitext(file)[0] + ext

r = sr.Recognizer()

for speaker in test_speakers:
    
    # create speaker folder if doesn't exist
    if not os.path.exists(os.path.join(predictions_dir, speaker)):
        os.makedirs(os.path.join(predictions_dir, speaker))
    print(f"speaker: {speaker}")

    if os.path.isdir(os.path.join(audio_dir, speaker)):
        for sentence in os.listdir(os.path.join(audio_dir, speaker)):
            if sentence.endswith(".wav"):
                audio_path = os.path.join(audio_dir, speaker, sentence)
                with sr.AudioFile(audio_path) as source:
                    audio_data = r.record(source)
                decoded = r.recognize_google(audio_data)
                out_file = replace_extension(os.path.join(predictions_dir, speaker, sentence), '.pred')
                f = open(out_file, 'w')
                f.write(decoded)
                f.close()
```

```python
from jiwer import wer

ground_truth = []
hypothesis = []

# preds_dir = './librispeech-big-lab-preds'
# labels_dir = './librispeech-big-lab'
preds_dir = './augmented-data-big-preds'
labels_dir = './augmented-data-big'
test_speakers = ['78', '302', '248']

def replace_extension(file, ext):
    return os.path.splitext(file)[0] + ext

for speaker in test_speakers:
    if os.path.isdir(os.path.join(preds_dir, speaker)):
        for pred in os.listdir(os.path.join(preds_dir, speaker)):
            if pred.endswith(".pred"):
                with open(replace_extension(os.path.join(labels_dir, speaker, pred), ".lab")) as f:
                    ground_truth.append(f.readline().lower())
                with open(os.path.join(preds_dir, speaker, pred)) as f:
                    hypothesis.append(f.readline().lower())
                
error2 = wer(ground_truth, hypothesis)
```
